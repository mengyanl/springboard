Part 1 ‚Äê Exploratory data analysis
The attached logins.json file contains (simulated) timestamps of user logins in a particular

```python
import json
import pandas as pd


# Opening JSON file
f = open('logins.json', 'r') 
# returns JSON object as a dictionary
data = pd.DataFrame(json.load(f))

f.close()

data.head()
```

```python
sorted_data = data.sort_values(by=['login_time'])
sorted_data.tail()
```

```python
import pandas as pd

sorted_data.index = pd.to_datetime(sorted_data.login_time.astype(str))
result = sorted_data.resample('15min').count()
result.column = 'login_count'
result.head(20)
```

## Part 2. Experiment and metrics design
The neighboring cities of Gotham and Metropolis have complementary circadian rhythms: on
weekdays, Ultimate Gotham is most active at night, and Ultimate Metropolis is most active
during the day. On weekends, there is reasonable activity in both cities.
However, a toll bridge, with a two way toll, between the two cities causes driver partners to tend
to be exclusive to each city. The Ultimate managers of city operations for the two cities have

    <li>b) what statistical test(s) you will conduct to verify the significance of the

## Answers:
<ol>
<li>I would choose profit increase (or decrease) after the experiment as the key measure of success, this is the most straight forward measurement and would be the most intestested measurement?</li>
<li>Experiment design: </li>
<ul>
    <li>a) implementation of the experiment:</li> 
    <ul>
       <li> if the number of drivers are above 500 randomly choose 5% - 10% of drivers form each cities </li>
       <li> if the number of drivers are above 500 randomly choose 10% - 20% of drivers form each cities</li>
       <li> if the number of drivers are bellow 100 randomly choose 50% drivers form each cities </li> 
       let the drivers know about the policy and reimburse the toll fees. Randomly select counter part drivers (randomly select drivers to match the numbers in the test group) as the null group. Record the activities of both group: what city they have been in during different time of the day or weekdays/weekends, how much they spend for the toll bridge, calculate the net profit from both groups. 
    </ul>
    <li>b) A student t-test (A/B test) on the net profits from each car from both the experimental and null group, a p-value of 0.05 shows significance. </li>
    <li>c) If the results are significant, the policy could be taken on all drivers. But before bring in the policy, the actual activities of drivers of both groups need to be examined carefully, making sure the increase of net profit is caused by the policy and the the toll bridge fee and other possible expenses have been excluded from the net profit. </li> 
    </ul>

```python
X = retention_data.drop(['signup_date', 'last_trip_date', 'retained', 'Month','6th_month_activiity'], axis = 1)
X.head()             
```

```python
# Seperate the column_names into numerical and categorical lists: cat and num
column_name = X.columns.values.tolist()
print('Column names:', column_name)
cat = []
num = []
for c in column_name:
    if X[c].dtype == 'object' or X[c].dtype == 'bool':
        cat.append(c)
    if X[c].dtype == 'int64' or X[c].dtype == 'float64':
        num.append(c)
print('Total of ', len(cat), 'Categorical columns:', cat) 
print('Total of ', len(num),'Numerical columns:', num)
```

```python
X_cat = X[cat]
dummies_X_cat = pd.get_dummies(X_cat,columns = cat , drop_first = True)
print(dummies_X_cat.shape)
dummies_X_cat.head()
```

```python
X_num = X[num]
X_all = dummies_X_cat.join(X_num)
X_all.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
from sklearn import preprocessing
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings('ignore')
```

```python
# check version number
import imblearn
print(imblearn.__version__)
from imblearn.over_sampling import SMOTE
from collections import Counter
```

```python
from sklearn.metrics import f1_score
```

```python
def fit_logreg_model(X,y,c):
    over = SMOTE(sampling_strategy=c)
    X, y = over.fit_resample(X, y)
    counter = Counter(y)

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
    
    scaler = preprocessing.MinMaxScaler().fit(X_train)
    X_train_scaled=scaler.transform(X_train)
    X_test_scaled=scaler.transform(X_test)

    logreg = LogisticRegression()
    logreg.fit(X_train,y_train)
    y_pred = logreg.predict(X_test)
    return [c, counter, f1_score(y_test, y_pred, average=None), f1_score(y_test, y_pred, average='macro')]
```

```python
X = X_all
y = retention_data['6th_month_activiity']
for c in [0.3,0.4,0.5,0.6,0.7,0.8,0.9,1]:
    print(fit_logreg_model(X,y,c))
```

```python
def fit_RF_model(X,y,c):
    over = SMOTE(sampling_strategy=c)
    X, y = over.fit_resample(X, y)
    counter = Counter(y)
    print('c: ', c, ', Counter: ', counter)
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
    
    scaler = preprocessing.MinMaxScaler().fit(X_train)
    X_train_scaled=scaler.transform(X_train)
    X_test_scaled=scaler.transform(X_test)

    RF = RandomForestClassifier(max_depth=3, random_state =42)
    RF.fit(X_train,y_train)
    y_pred = RF.predict(X_test)
 
    print('f1_none: ', f1_score(y_test, y_pred, average=None))
    print('f1_macro: ', f1_score(y_test, y_pred, average='macro'))
```

```python
X = X_all
y = retention_data['6th_month_activiity']
for c in [0.3,0.4,0.5,0.6,0.7,0.8,0.9,1]:
    print(fit_RF_model(X,y,c))
```

```python
X = X_all
y = retention_data['6th_month_activiity']

over = SMOTE(sampling_strategy=1)
X, y = over.fit_resample(X, y)
print(Counter(y))

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
    
scaler = preprocessing.MinMaxScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
X_train_scaled[0:1]

RF = RandomForestClassifier(max_depth=3, random_state = 1)
RF.fit(X_train,y_train)
y_pred = RF.predict(X_test)

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
# Searching for better hyperparameters
# Define Parameters
max_depth=[16, 32, 64]
n_estimators = [64, 256, 512]
param_grid = {'max_depth': max_depth, 
              'n_estimators': n_estimators}

# Build the grid search
dfrst = RandomForestClassifier(n_estimators=n_estimators, max_depth=max_depth, random_state = 1)
grid = GridSearchCV(estimator=dfrst, param_grid=param_grid, cv = 5, scoring = 'accuracy', verbose = 3)
grid_results = grid.fit(X_train, y_train)

# Summarize the results in a readable format
print("Best: {0}, using {1}".format(grid_results.cv_results_['mean_test_score'], grid_results.best_params_))
results_df = pd.DataFrame(grid_results.cv_results_)
results_df
```

```python
# Fit RandomForest model using tuned hyperparameters
RF = RandomForestClassifier(n_estimators= 64 , max_depth =  512 )
RF.fit(X_train,y_train)
y_pred = RF.predict(X_test)

# Metrics calculation
print(classification_report(y_test, y_pred))

# Create a confusion matrix
cnf_matrix = confusion_matrix(y_test, y_pred)
print(cnf_matrix)
```

```python
# Plot feature_importance for powerpoint use
feature_importance = RF.feature_importances_
print(feature_importance)
# make importances relative to max importance
feature_importance = 100.0 * (feature_importance / feature_importance.max())
sorted_idx = np.argsort(feature_importance)

pos = np.arange(sorted_idx.shape[0])
print(pos.size)
sorted_idx.size
plt.figure(figsize=(10,6))
plt.barh(pos[-10:], feature_importance[sorted_idx][-10:], align='center')
plt.yticks(pos[-10:], X.columns[sorted_idx][-10:])
plt.xlabel('Relative Importance')
plt.title('Variable Importance')
plt.show()
```
